# Normas

## Qué es una norma y para qué se usa. Desigualdad Triangular

La norma es el tamaño de un vector, siempre es mayor o igual a 0.

La función norma recibe de entrada un vector y a este se le asocia un número. Ese número es el tamaño del vector.

¿Porque querriamos medir el tamaño de un vector en Machine Learning? Porque querriamos saber cual es el error que estamos cometiendo al hacer nuestras aproximaciones o clasificaciones.

### Propiedades

1. La norma de un vector es 0, si y solo si el vector es 0.
2. **Desigualdad triangular**: La suma de dos vectores V1 y V2 nos da un vector V3. La norma del vector resultante V3 es menor o igual a la suma de las normas de V1 y V2.
3. La norma de un escalar por el vector es el valor absoluto del escalar por la norma del vector original.

Demostremos la propiedad de desigualdad triangular de fomra gráfica.

```
v1v2 = v1 + v2
print(v1v2)
```

Con el método **.norm(VECTOR)** del módulo de **algebra lineal** de **numpy** podemos saber la norma de un vector.

```
np.linalg.norm(v1v2)
```

## Tipos de normas

### Norma L0

* **La norma L0**: devuelve la cantidad de elementos diferentes de 0 del vector.

```
vector = np.array([1, 2, 0, 5, 6, 0])
print(np.linalg.norm(vector, ord=0))
```

Ejemplo: Supongamos que en este vector estuvieramos representando las compras que ha realizado cada uno de nuestros usuarios. Si quisieramos calcular cuantos usuarios han realizado comprar, deberiamos calcularle la norma 0 a este vector. 

### Norma L1

* **La norma L1**: devuelve la suma de los valores absolutos de los elementos del vector.

```
vector = np.array([1, -1, 1, -1, 1])

print(np.linalg.norm(vector, ord=1))
```

### Norma L2

* **La norma L2**: La magnitud del vector desde su origen. Distancia euclidiana. 

Esto tiene sentido hacerlo en R2 o R3, donde nuestra mente es capaz de ver una representación gráfica de nuestros vectores.

Los importante de las normas es que las podemos utilizar independientemente del tamaño del vector, del espacio en el cual estamos trabajando, aún cuando nosotros no logremos representar o ver cual sería su representación física.

```
vector = np.array([1, 1])

print(np.linalg.norm(vector))
print(np.linalg.norm(vector, ord=2))
```

Esto nos dice que la función norma por defecto calcula la norma de orden 2.

### Norma L2 al cuadrado 

* ¿Porque sería importante calcular la nomr L2²?

En machine learning se usa mucho la norma L2 pero elevada al cuadrado es decir, L2^2. **La ventaja de esto es computacional.**

* ¿Como podemos calcular L2² usando el producto interno? 

Vamos a definir un nuevo vector donde la intuicion no juegue un papel en la resolución de la norma de orden 2, o sea un vector con mas de 3 valores. 

```
vector = np.array([1, 2, 3, 4, 5, 6])
print(vector)
```

```
print(np.linalg.norm(vector, ord=2)**2)
```

```
print(vector.T.dot(vector))
```

Podríamos reunir muchos vectores en una matriz X, calcular X.dot(x_t) y tendríamos allí el cuadrado de las normas de todos esos vectores.


**Recordatorio**:  **¿por que al hacer el prducto interno se hace del vector transpuesto?**

Debido a que es producto de matrices, y se van a multiplicar vectores, lo que se hace es transponer para obtener las dimensiones correctas para la multiplicacion. O sea, el número de columnas del vector o matriz de la derecha tiene que ser igual al numero de filas de la matriz o vector de la izquierda.

Si se tiene dos vectores: 1xN y 1xN

* **1xN y 1xN NO se puede**
* Si se transpone uno de los dos quedaría 1xN y Nx1, entonces  **1xN y Nx1 si se puede**.


### Norma infinito


* **La norma infinito**: devuelve el valor absoluto más grande del vector.

```
vector = np.array([1, 2, 3, -100])

print(np.linalg.norm(vector, ord=np.inf))
```

El valor infinito en numpy es **np.inf**.

## El producto interno como función de una norma y su visualización

```
%matplotlib inline

import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```
v1 = np.array([0,0,0,3])
v2 = np.array([0,0,3,3])
```

```
plt.xlim(-2,6)
plt.ylim(-2,6)

plt.quiver([v1[0], v2[0]],
           [v1[1], v2[1]],
           [v1[2], v2[2]],
           [v1[3], v2[3]],
           angles='xy', scale_units='xy', scale=1,
           color=sns.color_palette())
plt.show()
```

En machine learniing se conoce como el coseno similitud entre dos vectores al coseno entre dos vectores. Entre más pequeño sea el angulo, más similitud habrá entre los vectores comparados.

La siguiente igualdad relaciona el producto escalar y el producto entre las normas de los vectores por el coseno del angulo que forman ambos.

v1^t * v2 = L2(v1) * L2(v2) * cos(a).

Comprobemos la formula anterior:

```
# v1^t * v2
v1 = np.array([0,3])
v2 = np.array([3,3])

print(v1.T.dot(v2))
```

```
# L2(v1) * L2(v2) * cos(a). 
norma_v1 = np.linalg.norm(v1)
norma_v2 = np.linalg.norm(v2)

print(norma_v1 * norma_v2 * np.cos(np.deg2rad(45)))
```

Si quisieramos obtener el angulo entre dos vectores que representen algo.

Definimos los vectores

```
v1 = np.array([0,3])
v2 = np.array([3,3])
```

Realizamos su producto interno.


```
prod_in = v1.T.dot(v2)
```

Calculamos las L2.

```
v1_norm = np.linalg.norm(v1)
v2_norm = np.linalg.norm(v2)
```

Para calcular el angulo nos aprovechamos de la igualdad de:
* v1^t * v2 = L2(v1) * L2(v2) * cos(a).

```
cos_a = prod_in / (v1_norm * v2_norm)
```

Despejamos el cos.

```
ang_rad_a = np.arccos(cos_a)
```

Pasamos de radiantes a grados.


```
ang_gra_a = np.rad2deg(ang_rad_a)
print(ang_gra_a)
```

 Nos devuelve 45, esto lo que nos dice es que hay entre 45º entre v1 y v2.

Esto es utilizado en machine learning para encontrar diferencias entre documentos o [qué tan parecidos son dos textos](https://www.youtube.com/watch?v=z0b7TAVtHrw) por ejemplo. Si evaluo una serie de oraciones como vectores y comparo con otra pueden pasar dos cosas: 
* Si el angulo que forman es muy pequeño podemos concluir que las dos cadenas de oraciones se parecen.
* Si por el contrario el angulo que forman es de 90 grados podemos decir que no se parecen en nada.
